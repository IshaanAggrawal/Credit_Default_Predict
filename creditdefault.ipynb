### Data Cleaning and Preprocessing

Before model training, the data is preprocessed. This involves checking for and handling missing values and duplicate rows. Additionally, the target variable is separated from features, and the dataset is split into training and testing sets. Crucially, SMOTE (Synthetic Minority Over-sampling Technique) is applied to address class imbalance in the training data, and `StandardScaler` is used to standardize features.

```python
from imblearn.over_sampling import SMOTE as smote
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.linear_model import LogisticRegression

y = df['default.payment.next.month']
x = df.drop(['default.payment.next.month', 'ID'], axis=1) # Drop both target and ID


x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42, stratify=y
)


model_pipeline = ImbPipeline(steps=[
    ('scaler', StandardScaler()),
    ('smote', smote(random_state=42)),
])

model_pipeline.fit(x_train, y_train)
```

### Regression Model Attempt (Demonstrative)

This section demonstrates that regression models are not suitable for a classification problem like predicting credit card default. Various regression models are applied, and their performance is evaluated using metrics like R2, MAPE, and a custom accuracy with tolerance, highlighting their limitations for this type of task.

**Used regression to demonstrate that it is a classification problem so results will be bad with regressions**:

```python
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import make_scorer

# Import Models
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

def custom_accuracy(y_true, y_pred, tolerance=0.10):
    y_true = np.where(y_true == 0, 1e-6, y_true)
    diff = np.abs((y_true - y_pred) / y_true)

    return np.mean(diff <= tolerance)
accuracy_scorer = make_scorer(custom_accuracy, tolerance=0.10)

regression_models = {
    "Linear Regression": LinearRegression(),
    "Ridge (L2)": Ridge(alpha=1.0),
    "Lasso (L1)": Lasso(alpha=0.1),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "SVR": SVR(kernel='rbf'),
    "KNN": KNeighborsRegressor(n_neighbors=5)
}

results = []
print(f"{'Model Name':<20} | {'R2':<8} | {'MAPE':<8} | {'Acc (10% Tol)':<15}")
print("-" * 65)
for name, model in regression_models.items():
    pipeline = Pipeline(steps=[
        ('scaler', StandardScaler()),
        ('regressor', model)
    ])

    r2_scores = cross_val_score(pipeline, x_train, y_train, cv=5, scoring='r2')
    mape_scores = cross_val_score(pipeline, x_train, y_train, cv=5, scoring='neg_mean_absolute_percentage_error')
    acc_scores = cross_val_score(pipeline, x_train, y_train, cv=5, scoring=accuracy_scorer)

    avg_r2 = r2_scores.mean()
    avg_mape = -mape_scores.mean()
    avg_acc = acc_scores.mean()

    print(f"{name:<20} | {avg_r2:.4f}   | {avg_mape:.2%}   | {avg_acc:.2%}")
    results.append({
        'Model': name,
        'R2': avg_r2,
        'MAPE': avg_mape,
        'Custom Accuracy': avg_acc
    })
```

### Deep Learning Model (Neural Network)

Here, a Sequential Keras model (a neural network) is built, compiled, and trained. It includes `Dense` layers with `relu` activation and `Dropout` for regularization. `SMOTE` is applied before training to handle class imbalance, and `StandardScaler` is used for feature scaling. The model's performance is evaluated using accuracy and a classification report.

```python
from sklearn.metrics import classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from imblearn.over_sampling import SMOTE # Import SMOTE
from sklearn.preprocessing import StandardScaler # Import StandardScaler
from sklearn.model_selection import train_test_split # Import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

smote = SMOTE(random_state=42)
x_train_res, y_train_res = smote.fit_resample(x_train, y_train)

model = Sequential()
model.add(Dense(32, activation='relu', input_dim=x_train_res.shape[1]))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

callback = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

model.fit(x_train_res, y_train_res, validation_data=(x_test, y_test), epochs=50, batch_size=32, callbacks=[callback])

loss, acc = model.evaluate(x_test, y_test)
print(f"Accuracy: {acc*100:.2f}%")

preds = (model.predict(x_test) > 0.5).astype(int)
print(classification_report(y_test, preds))
```

### XGBoost Model

This section trains and evaluates an XGBoost Classifier, a powerful gradient boosting algorithm. `scale_pos_weight` is used to handle class imbalance directly within the XGBoost model. The model's accuracy and a detailed classification report are printed to assess its performance.

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    scale_pos_weight=3
)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2%}")
print(classification_report(y_test, y_pred))
```

### Model Export

Finally, the trained XGBoost model and the `StandardScaler` object are saved to disk using `joblib`. This allows for the later deployment or reuse of the model and scaler without needing to retrain them.

```python
import joblib

model_export = model
joblib.dump(model_export, 'credit_default_model.pkl')

scaler_to_export = model_pipeline.named_steps['scaler']

joblib.dump(scaler_to_export, 'scaler.pkl')
```
